```python
from power_calcs import power_calc
import numpy as np
import matplotlib.pyplot as plt

# plot a range (output) of areas needed to achieve 400W for an input range of velocities
velocities = np.linspace(0.5,2.6,20)
areas = np.zeros(len(velocities))
efficiencies = np.linspace(0.1,0.5,5)

for n in efficiencies:

    for i,v in enumerate(velocities):
        p, e, r, areas[i], v = power_calc(P=400, efficiency=n, V=v)

    plt.plot(velocities, areas, label='Efficiency = {}'.format(n))

# draw a vertical line at the point where y = 1
x = np.linspace(0,2.6,20)
y = np.ones(len(x))
plt.plot(x,y,'r--')

# draw a vertical line at the 20% velocity given the distribution is 0.14 to 0.26
y = np.linspace(0,10,20)
x = np.ones(len(x))
plt.plot(x,y,'g--')

plt.legend()

plt.xlabel('Velocity (m/s)')
plt.ylabel('Minimum Area (m^2)')
plt.show()
```
