```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# list of months for x-axis for 10 years
months = ['Year1','Year2','Year3','Year4','Year5','Year6','Year7','Year8','Year9','Year10']

payg = np.array([-423812.08, -274816.27,-125820.45, -400636.72, -251640.90, -102645.08, -377461.35, -228465.53, -79469.71, -69526.10 ])
ngo =  np.array([75000.00 ,75000.00, 75000.00 ,150000.00 ,150000.00 ,150000.00 ,225000.00 ,225000.00 ,225000.00 ,300000.00 ])
total = payg + ngo

# plot the data on a single line graph
plt.figure(figsize=(10,5))
plt.plot(months, payg, label='PAYG balance', color='r', ls='--', alpha=0.8)
plt.plot(months, ngo, label='NGO balance', color='b', ls='-.', alpha=0.8)
plt.plot(months, total, label='Total', color='g', ls='-')
plt.hlines(0,0,9, colors='k', alpha=0.4)
# mark the break-even point
plt.annotate('Break-even point', xy=(7,0), xytext=(4.5,250000), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=22)

# make x ticks in 1000s of dollars
plt.yticks(np.arange(-500000, 500000, 100000), labels=[f'${i/1000}K' for i in np.arange(-500000, 500000, 100000)])

# mark the downward section as production cost
plt.annotate('Repayments', xy=(1,-270000), xytext=(0,-90000), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=15)

plt.annotate('New purchase', xy=(3,-400000), xytext=(4,-450000), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=15)

# plt.xlabel('Months', fontsize=15)
plt.ylabel('Balance', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# plt.title('PAYG vs NGO sales for 10 years')
plt.legend(fontsize=15)
plt.grid()
plt.show()
```
